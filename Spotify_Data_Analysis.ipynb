# Importing Libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# Loading the Datasets

```python
tracks_df = pd.read_csv("tracks.csv")
```

# Top 5 rows

```python
tracks_df.head()
```

# Analysing the Dataset

```python
pd.isnull(tracks_df).sum()
```

```python
tracks_df.info()
```

```python
tracks_df.describe().transpose()
```

# Top 5 Least Popular Songs

```python
sorted_tracks = tracks_df.sort_values("popularity", ascending = True)
sorted_tracks.head()
```

# Top 5 Most Popular Tracks

```python
most_popular = tracks_df.query("popularity>95",inplace=False)
most_popular.sort_values("popularity",ascending=False).head()
```

# Reconsidering the Index 

```python
tracks_df['release_date'] = pd.to_datetime(tracks_df['release_date'], errors='coerce')
tracks_df.set_index("release_date", inplace=True)
tracks_df.head()
```

# Converting duration in ms to seconds

```python
tracks_df["duration"] = tracks_df["duration_ms"].apply(lambda x: round(x/1000))
tracks_df.drop("duration_ms",inplace=True,axis=1)
```

```python
tracks_df["duration"].head()
```

# Visualisation

```python
numeric_df = tracks_df.drop(["key", "mode", "explicit"], axis=1).select_dtypes(include=[float, int])
```

```python
corr_df = numeric_df.corr(method="pearson")
```

```python
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".2f", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmap between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
sample_df = tracks_df.sample(int(0.004*len(tracks_df)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="c").set(title="Loudness vs Energy")
```

```python
tracks_df["dates"] = tracks_df.index.get_level_values("release_date")
tracks_df["dates"] = pd.to_datetime(tracks_df["dates"])
years = tracks_df["dates"].dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
avg_duration_per_year = tracks_df.groupby(years)['duration'].mean().reset_index()

fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x=avg_duration_per_year['dates'], y=avg_duration_per_year['duration'], ax=ax)
ax.set_title("Year vs Average Duration")
plt.xticks(rotation=90)
plt.show()
```

```python
total_duration_per_year = tracks_df.groupby(years)['duration'].sum().reset_index()

sns.set_style(style="whitegrid")
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize=fig_dims)
sns.lineplot(x=total_duration_per_year['dates'], y=total_duration_per_year['duration'], ax=ax).set(title="Year vs Total Duration")
plt.xticks(rotation=90)
plt.show()
```

```python
features_df = pd.read_csv("SpotifyFeatures.csv")
features_df.head()
```

```python
plt.title("Duration of the Songs in different genres")
sns.barplot(y="genre",x="duration_ms",data=features_df,palette="rocket",hue="genre")
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```
